```python
# Importing necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
# Load the dataset with specified encoding
data = pd.read_csv('sales_data_sample.csv', encoding='ISO-8859-1')


# Explore the dataset (Optional, just to understand the data structure)
# print(data.head())
# print(data.columns)

# Selecting relevant numerical columns for clustering
# For demonstration, let's assume we want to cluster based on 'SALES' and 'QUANTITYORDERED'
# You may need to preprocess the dataset based on what columns are available and relevant.
X = data[['SALES', 'QUANTITYORDERED']]

# Scaling the data (important for clustering algorithms)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method to determine the optimal number of clusters
sse = []  # List to store the sum of squared errors for each value of k
k_range = range(1, 11)  # Testing k values from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)  # Inertia is the sum of squared distances to nearest cluster center

# Plotting the Elbow graph
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```
